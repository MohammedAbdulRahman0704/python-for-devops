### Datastructures:Lists and Dictionaries

#### List

```python
fruits = ["apple", "pear", "strawberry"]
type(fruits)
```

```python
fruits[:-1]
```

```python
for my_fruits in fruits:
  print(f"This is my favorite fruit: {my_fruits}")
```

#### Dictionary

```python
meal = {"Breakfast": "Black Coffee", "Lunch" : "Salad", "Dinner": "Rice and Chicken"}
```

```python
meal.keys()
```

```python
meal.values()
```

```python
for meal_name, meal_item in meal.items():
  print(f"This meal {meal_name} is when I like to consume {meal_item}")
```
